```python
d1 = {
      'a': 100 ,
      'b': 200,
      'c': 300
      }

d2 = {
       'a':300,
       'b':200,
       'd':400 
      }
d3 ={}
for key1 , key2 in zip(d1,d2) :
    if key1 == key2:
        d3[key1]= d2[key2] + d1[key1]
    elif key1 != key2 :
        d1_new_key= key1
        d2_new_key= key2
        d3[d1_new_key] = d1.get(key1) 
        d3[d2_new_key] = d2.get(key2) 



d3 
   
```
